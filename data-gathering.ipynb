```python
import os
import json
import sys
import spotipy
from spotipy.oauth2 import SpotifyOAuth
from dotenv import load_dotenv



# Load environment variables from .env file
load_dotenv()


CLIENT_ID = os.getenv('SPOTIPY_CLIENT_ID')
CLIENT_SECRET = os.getenv('SPOTIPY_CLIENT_SECRET')
REDIRECT_URI = os.getenv('SPOTIPY_REDIRECT_URI')
TARGET_PLAYLIST_NAME = 'Top songs 2024'

if not all([CLIENT_ID, CLIENT_SECRET, REDIRECT_URI]):
    sys.exit("Missing Spotify credentials. Please check your .env file.")

SCOPE = 'user-library-read user-top-read'

# Authenticate with Spotify
sp_oath = spotipy.Spotify(auth_manager=SpotifyOAuth(
    client_id=CLIENT_ID,
    client_secret=CLIENT_SECRET,
    redirect_uri=REDIRECT_URI,
    scope=SCOPE
))

def get_user_playlist_by_name(playlist_name, sp_client):
    """Search the current user's playlists by name (case-insensitive)"""
    playlists = []
    results = sp_client.current_user_playlists()
    playlists.extend(results.get('items', []))
    while results.get('next'):
        results = sp_client.next(results)
        playlists.extend(results.get('items', []))
    
    for playlist in playlists:
        if playlist['name'].lower() == playlist_name.lower():
            return playlist
    return None

def search_playlist_globally(playlist_name, sp_client):
    """Use Spotify's search endpoint to find a playlist by name."""
    results = sp_client.search(q=playlist_name, type='playlist')
    playlists = results.get('playlists', {}).get('items', [])
    for playlist in playlists:
        if playlist['name'].lower() == playlist_name.lower():
            return playlist
    return None

def get_all_playlist_tracks(playlist_id, sp_client):
    """Fetch all tracks from a given playlist, handling pagination."""
    results = sp_client.playlist_tracks(playlist_id)
    tracks = results.get('items', [])
    while results.get('next'):
        results = sp_client.next(results)
        tracks.extend(results.get('items', []))
    return tracks

```

```python
import os
import requests
from dotenv import load_dotenv

def get_access_token():
    load_dotenv()

    client_id = os.getenv("SPOTIPY_CLIENT_ID")
    client_secret = os.getenv("SPOTIPY_CLIENT_SECRET")

    # Get the token using Spotify's API
    token_url = "https://accounts.spotify.com/api/token"
    headers = {
        "Content-Type": "application/x-www-form-urlencoded"
    }
    data = {
        "grant_type": "client_credentials",
        "client_id": client_id,
        "client_secret": client_secret
    }
    
    response = requests.post(token_url, headers=headers, data=data)

    if response.status_code == 200:
        return response.json()["access_token"]
    else:
        raise Exception("Failed to get access token: " + response.text)
```
